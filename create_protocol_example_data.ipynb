```r
# Experimental Task 1
In this experimental task, we modified the dataset "/mnt/mfs/ctcn/datasets/rosmap/tmt/dlpfcTissue/round_1_2_harmonized/combine_r1andr2_protein.reg_cov_cog.uniq.proj.csv" to get a new simulation dataset which can be used to repeat our pQTL experiment.
```

```r
### Create folder path for saving dataset got from experimental result
```

```r
mkdir -p /mnt/mfs/hgrcgrid/homes/xx2404/xQTL_output/newdata/uniqProjcsv
```

```r
### Input
```

```r
# read the raw data
data <- read.csv("/mnt/mfs/hgrcgrid/homes/zq2209/combine_r1andr2_protein.reg_cov_cog.uniq.proj.csv")
head(data)
dim(data)
```

```r
### Modify column name in raw data
```

```r
# change the sample name in the column of the raw data
for (i in 2:length(data)){
    string <- "sample"
    number <- i-1
    colnames(data)[i] <- paste0(string, sep = "_", number)}
head(data)
dim(data)
```

```r
### Delete some rows which have NA values
This is a flexible selection depended on experiment condition. As for the NA question, there are many different approaches to 
deal with. In this experimental task, it is not a mandatory requirement to settle the NA issue. A simple but effective method of solving the NA question is given out as following:
```

```r
# Search for the NA in each row, if find NA in a row, then delete the row
 rows_with_na <- apply(data, 1, function(row) any(is.na(row)))
 data <- data[!rows_with_na, ]
 head(data)
 dim(data)
```

```r
 # Check whether there are still some NAs in the data
 has_na <- any(is.na(data))
 has_na

 # If find there are some NAs in the data, display the locations of the NAs in the data
 NA_indices <- which(is.na(data)=="TRUE")
 NA_indices 
```

```r
### Add some random noise to raw data for constructing new dataset
In this step, we add some random noise factors into the raw dataset for obtaining a new similar simulation dataset which can be distinguished from the raw data. Notely, this step is not mandatory requirement. The random noise factor makes the new dataset be similar to the raw dataset but not the same.
```

```r

#set random number seed to make sure the result can be repeated
set.seed(123)

n_rows <- nrow(data)
n_cols <- ncol(data)

# add random noise for the raw data
noise_factor <- 0.1

for (i in 1:n_rows){
    for (j in 2:n_cols){
        data[i,j] <- data[i,j] + runif(1, -noise_factor, noise_factor)
        }
    }
head(data)
dim(data)
```

```r
### Save the dataset after some processes
```

```r
# Save the data that was implemented several processes
write.csv(data, file = "/mnt/mfs/hgrcgrid/homes/xx2404/xQTL_output/newdata/uniqProjcsv/processed_data.csv", row.names=FALSE)
```

```r
# Experimental Task 2
In experimental task 2, we extract some partial information from the dataset "/mnt/mfs/ctcn/datasets/rosmap/phenotypes/2022Feb08/dataset_707_basic_02-08-2022.clean.txt" and modify these partial information to get a new simulation dataset which can be used to repeat our follow-up experiment.
```

```r
### Input
This dataset comes from the "covariates_preprocessing" script
```

```r
# read the raw dataset
data <- read.table("/mnt/mfs/ctcn/datasets/rosmap/phenotypes/2022Feb08/dataset_707_basic_02-08-2022.clean.txt", header = TRUE, sep = '\t')
head(data)
dim(data)
```

```r
### Extract the "sex" and "age" columns from the raw dataset 
```

```r
# Extract the "msex" and "age_death" columns and put these two columns together
sex_index <- which(colnames(data)=="msex")
agedeath_index <- which(colnames(data)=="age_death")
data <- data[,c(sex_index, agedeath_index)]
head(data)
```

```r
### Numerical integer
```

```r
# Make the float values corresponding to the "age_death" column be integer values
data$age_death <- round(data$age_death, 0)
new_col_names <- c("sex", "age")
names(data) <- new_col_names
head(data)
```

```r
### Dataset save
Save the dataset obtained by extracting two important columns and integrating them together, the final dataset format continue to keep be .txt as the raw dataset.
```

```r
write.table(data, file = "/mnt/mfs/hgrcgrid/homes/xx2404/xQTL_output/newdata/sampleSheet/processed_data.txt", sep = "\t", row.names = FALSE)
```
